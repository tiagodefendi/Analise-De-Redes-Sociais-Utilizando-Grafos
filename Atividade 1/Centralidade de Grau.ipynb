```python
# Importação de bibliotecas
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

print("Bibliotecas importadas com sucesso!")
```

```python
# Definição de um grafo não orientado de exemplo
G_nao_orientado = nx.Graph()

# Adicionando nós
G_nao_orientado.add_nodes_from([1, 2, 3, 4, 5])

# Adicionando arestas - criando uma estrutura em estrela onde o nó 1 é central
arestas_nao_orientado = [(1, 2), (1, 3), (1, 4), (1, 5), (2, 3), (4, 5)]
G_nao_orientado.add_edges_from(arestas_nao_orientado)

print("Grafo não orientado criado:")
print("Nós:", list(G_nao_orientado.nodes()))
print("Arestas:", list(G_nao_orientado.edges()))
```

```python
# Definição de um grafo orientado de exemplo
G_orientado = nx.DiGraph()

# Adicionando nós
G_orientado.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

# Adicionando arestas direcionadas - criando uma estrutura hierárquica
arestas_orientado = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E'), ('E', 'A')]
G_orientado.add_edges_from(arestas_orientado)

print("Grafo orientado criado:")
print("Nós:", list(G_orientado.nodes()))
print("Arestas:", list(G_orientado.edges()))
```

```python
# Cálculo das centralidades de grau

# Para o grafo não orientado
degree_nao_orientado = nx.degree_centrality(G_nao_orientado)

# Para o grafo orientado
in_degree_orientado = nx.in_degree_centrality(G_orientado)
out_degree_orientado = nx.out_degree_centrality(G_orientado)

print("Centralidades de grau para o grafo não orientado:")
for no, centralidade in degree_nao_orientado.items():
    print(f"Nó {no}: {centralidade:.3f}")

print("\nCentralidades de grau de entrada para o grafo orientado:")
for no, centralidade in in_degree_orientado.items():
    print(f"Nó {no}: {centralidade:.3f}")

print("\nCentralidades de grau de saída para o grafo orientado:")
for no, centralidade in out_degree_orientado.items():
    print(f"Nó {no}: {centralidade:.3f}")
```

```python
# Visualização do grafo não orientado com centralidade de grau

# Subplot para o grafo não orientado
plt.figure(figsize=(10, 5))
ax1 = plt.subplot(121)
pos_nao_orientado = nx.spring_layout(G_nao_orientado)
cores_nao_orientado = [degree_nao_orientado[no] for no in G_nao_orientado.nodes()]

nx.draw_networkx_nodes(G_nao_orientado, pos_nao_orientado,
                       node_color=cores_nao_orientado,
                       cmap=plt.cm.plasma,
                       node_size=500,
                       ax=ax1) # Specify the axes
nx.draw_networkx_edges(G_nao_orientado, pos_nao_orientado, ax=ax1) # Specify the axes
nx.draw_networkx_labels(G_nao_orientado, pos_nao_orientado, ax=ax1) # Specify the axes
plt.title('Grafo Não Orientado\nCentralidade de Grau')
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.plasma,
                                  norm=plt.Normalize(vmin=min(cores_nao_orientado),
                                                    vmax=max(cores_nao_orientado))),
             ax=ax1, shrink=0.7) # Specify the axes

# Subplot para o grafo orientado
ax2 = plt.subplot(122)
pos_orientado = nx.spring_layout(G_orientado)
cores_in = [in_degree_orientado[no] for no in G_orientado.nodes()]
cores_out = [out_degree_orientado[no] for no in G_orientado.nodes()]

# Visualizando centralidade de entrada
nx.draw_networkx_nodes(G_orientado, pos_orientado,
                       node_color=cores_in,
                       cmap=plt.cm.plasma,
                       node_size=500,
                       ax=ax2) # Specify the axes
nx.draw_networkx_edges(G_orientado, pos_orientado,
                       arrowstyle='->', arrowsize=15, ax=ax2) # Specify the axes
nx.draw_networkx_labels(G_orientado, pos_orientado, ax=ax2) # Specify the axes
plt.title('Grafo Orientado\nCentralidade de Grau de Entrada')
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.plasma,
                                  norm=plt.Normalize(vmin=min(cores_in),
                                                    vmax=max(cores_in))),
             ax=ax2, shrink=0.7) # Specify the axes

plt.tight_layout()
plt.show()
```

```python
# Visualização adicional mostrando a centralidade de saída no grafo orientado
plt.figure(figsize=(6, 5))
nx.draw_networkx_nodes(G_orientado, pos_orientado,
                       node_color=cores_out,
                       cmap=plt.cm.plasma,
                       node_size=500)
nx.draw_networkx_edges(G_orientado, pos_orientado,
                       arrowstyle='->', arrowsize=15)
nx.draw_networkx_labels(G_orientado, pos_orientado)
plt.title('Grafo Orientado\nCentralidade de Grau de Saída')
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.plasma,
                                  norm=plt.Normalize(vmin=min(cores_out),
                                                    vmax=max(cores_out))),
             ax=plt.gca(),  # Specify the current axes
             shrink=0.7)
plt.show()
```

```python
# Carregar o dataset Zachary's Karate Club
G_karate = nx.karate_club_graph()
```

```python
# Examinar as propriedades básicas da rede (número de nós, arestas, se é dirigida)

print("Grafo Zachary's Karate Club:")
print("Nós:", list(G_karate.nodes()))
print("Arestas:", list(G_karate.edges()))
print(f"{'É' if G_karate.is_directed() else 'Não é'} dirigido")
```

```python
# Centralidade de Grau (Degree Centrality)
degree_karate = nx.degree_centrality(G_karate)
for no, centralidade in degree_karate.items():
    print(f"Nó {no}: {centralidade:.3f}")
```

```python
# Criar visualizações da rede, colorindo os nós de acordo com cada medida de centralidade
# Identificar os 5 nós mais centrais para cada tipo de centralidade
# Comparar os resultados das diferentes medidas de centralidade

# plot para o grafo não orientado
plt.figure(figsize=(10, 5))
ax1 = plt.subplot(121)
pos_karate = nx.spring_layout(G_karate)
cores_karate = [degree_karate[no] for no in G_karate.nodes()]

nx.draw_networkx_nodes(G_karate, pos_karate,
                       node_color=cores_karate,
                       cmap=plt.cm.plasma,
                       node_size=300,
                       ax=ax1) # Specify the axes
nx.draw_networkx_edges(G_karate, pos_karate, ax=ax1) # Specify the axes
nx.draw_networkx_labels(G_karate, pos_karate, ax=ax1) # Specify the axes
plt.title('Grafo Karate\nCentralidade de Grau')
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.plasma,
                                  norm=plt.Normalize(vmin=min(cores_karate),
                                                    vmax=max(cores_karate))),
             ax=ax1, shrink=0.8) # Specify the axes
```

```python
# Centralidade de Intermediação (Betweenness Centrality)
betweenness_karate = nx.betweenness_centrality(G_karate)
for no, centralidade in betweenness_karate.items():
    print(f"Nó {no}: {centralidade:.3f}")
```

```python
# Centralidade de Proximidade (Closeness Centrality)
closeness_karate = nx.closeness_centrality(G_karate)
for no, centralidade in closeness_karate.items():
    print(f"Nó {no}: {centralidade:.3f}")
```

```python
# Centralidade de Autovetor (Eigenvector Centrality)
eigenvector_karate = nx.eigenvector_centrality(G_karate)
for no, centralidade in eigenvector_karate.items():
    print(f"Nó {no}: {centralidade:.3f}")
```
